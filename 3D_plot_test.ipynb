```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
```

```python
s25 = pd.read_csv('../spectrafit/300 C/FA_3.6wt__300C_25s.csv', header=None, names=('wavenumber', 'counts'))
s35 = pd.read_csv('../spectrafit/300 C/FA_3.6wt__300C_35s.csv', header=None, names=('wavenumber', 'counts'))
s45 = pd.read_csv('../spectrafit/300 C/FA_3.6wt__300C_45s.csv', header=None, names=('wavenumber', 'counts'))
s55 = pd.read_csv('../spectrafit/300 C/FA_3.6wt__300C_55s.csv', header=None, names=('wavenumber', 'counts'))
s65 = pd.read_csv('../spectrafit/300 C/FA_3.6wt__300C_65s.csv', header=None, names=('wavenumber', 'counts'))
```

```python
fig = plt.figure(figsize=(15,6))
ax = fig.add_subplot(111, projection='3d')
ax.plot(s25['wavenumber'], s25['counts'], zs=25, zdir='y', c='blue', linewidth=0.75)
ax.plot(s35['wavenumber'], s35['counts'], zs=35, zdir='y', c='blue', linewidth=0.75)
ax.plot(s45['wavenumber'], s45['counts'], zs=45, zdir='y', c='blue', linewidth=0.75)
ax.plot(s55['wavenumber'], s55['counts'], zs=55, zdir='y', c='blue', linewidth=0.75)
ax.plot(s65['wavenumber'], s65['counts'], zs=65, zdir='y', c='blue', linewidth=0.75)

ax.view_init(30, -100)
ax.set_xlabel('wavenumber')
ax.set_ylabel('resonance time')
ax.set_zlabel('counts')
```
